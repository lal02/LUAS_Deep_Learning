However, I probably have to grade this again later, since I believe there's quite a lot of problems with the categorical handling. For example "Color" seems to be replaced with numbers.

However, Color is a nominal category, which means, we can't replace colors with various numbers, like Yellow = 1, Blue = 2, Green = 3 etc. since it doesn't mean anything statistically. E.g. we can't imply "Green" is "more" than "Yellow", since they are not comparable, that's why 1 and 3 doesn't work here.

We have to create a separate binary variable for each color option, like Yellow (0/1), Blue (0/1), Green (0/1) etc. 

See the "Common pitfalls" material in Moodle, there's an explanation on this smile

Check also the other categorical variables in your dataset so that you don't accidentally replace them with just a random number.

Best way on how to convert a nominal category into binary variables, check lecture examples, Housing dataset.

Solution:
=> Change Color, Category to boolean flags

```python
# this makes multiple columns with the variable (Separate for yes/no)
from sklearn.preprocessing import OneHotEncoder
#variables = ['Category','Color']
variables = ['Category']

# use encoder
encoder = OneHotEncoder(sparse_output=False).set_output(transform="pandas")
one_hot_encoded = encoder.fit_transform(df[variables]).astype(int)
df = pd.concat([df,one_hot_encoded],axis=1).drop(columns=variables)

df.to_csv("../datasets/Task_1/8.csv")
```
