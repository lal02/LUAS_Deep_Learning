# Final Conclusion:
After a lot of preprocessing and a lot of optimization the model is at a point where it might be usable but still not good.
This is definitely not a synthetic dataset and was filled with quite a few problems where I had to find workarounds first.
Car prizes in itself are probably not easy to predict even with a perfect model due to the complexity of the topic. 
E.g. some used cars are way more expensive than other cars of the same category because in hindsight it was a really good manufactured car. Factors like this can easily confuse the model. 

The current status of the model seems to undershoot a lot and predict too little price, while getting medium to low priced cars correct.
This might be because there are too little high priced cars or too many low priced cars or an entirely different reason
